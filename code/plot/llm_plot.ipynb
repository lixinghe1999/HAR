```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator

# 
precision_file_path = '/home/bufang/Qwen2-Audio/main/results_ego4d_50cls_new/Precision.txt'
recall_file_path = '/home/bufang/Qwen2-Audio/main/results_ego4d_50cls_new/Recall.txt'
f1_file_path = '/home/bufang/Qwen2-Audio/main/results_ego4d_50cls_new/F1_Score.txt'

#  Precision  P
with open(precision_file_path, 'r') as f:
    P = [float(line.strip()) for line in f.readlines()]

#  Recall  R
with open(recall_file_path, 'r') as f:
    R = [float(line.strip()) for line in f.readlines()]

#  F1_Score  F1
with open(f1_file_path, 'r') as f:
    F1 = [float(line.strip()) for line in f.readlines()]

#  P, R, F1 
assert len(P) == len(R) == len(F1), "P, R, F1 "

# 
fig, ax = plt.subplots(figsize=(11, 3))
bar_width = 0.2  # 
x = list(range(len(F1)))  # x 

#  PrecisionRecall  F1 
plt.bar([i for i in x], P, width=bar_width, color='lightskyblue', hatch='//', label='Precision')
plt.bar([i + bar_width for i in x], R, width=bar_width, color='lightcoral', hatch='--', label='Recall')
plt.bar([i + 2 * bar_width for i in x], F1, width=bar_width, color='lightgrey', hatch='/', label='F1-score')

#  x 
plt.xticks([i + bar_width for i in x],  # 
           [str(i + 1) for i in x],     #  1 
           fontsize=12)

#  y 
plt.yticks(fontsize=12)
ax.yaxis.set_major_locator(MultipleLocator(0.2))  # y  0.2

# 
plt.legend(loc='lower center', bbox_to_anchor=(0.5, 0.78), ncol=3, fontsize=12, handlelength=1)

# 
plt.ylabel('Score', fontsize=14)
plt.xlabel('Activity Index', fontsize=14)
plt.xlim(-0.5, len(x) - 0.5)  

# 
plt.tight_layout()

# 
plt.savefig('activities.pdf', bbox_inches='tight')
plt.show()
```
